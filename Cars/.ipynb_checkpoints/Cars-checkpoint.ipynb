```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Setting up the display of row and column to show all of the them
pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 50)
#loading all three data to dataframes using Pandas
dataframe1 = pd.read_csv("data1.csv")
dataframe2 = pd.read_csv("data2.csv")
dataframe3 = pd.read_csv("data3.csv")
```

```python
#Merged dataframe1 and dataframe2 based on 'id' column
combineddata = pd.merge(dataframe1, dataframe2, on = 'id')
finalcombineddata = pd.concat([combineddata, dataframe3])
finalcombineddata.to_csv("finalcombineddata.csv", index = False)
```

```python
#Replacing sixth with six
to_replace = {'sixth':'six'}
finalcombineddata.replace(to_replace, inplace=True)
```

```python
#Printing the duplicate rows
duplicate_rows_finalcombineddata = finalcombineddata[finalcombineddata.duplicated()]
print ("number of duplicate rows: ", duplicate_rows_finalcombineddata.shape)
print(duplicate_rows_finalcombineddata)
```

```python
duplicate_rows_finalcombineddata_id = finalcombineddata[finalcombineddata.duplicated(['id'])]
print (duplicate_rows_finalcombineddata_id.shape)
print(duplicate_rows_finalcombineddata_id)
```

```python
#Using drop_duplicates, the duplicate rows are dropped from the finalcombineddata
finalcombineddata = finalcombineddata.drop_duplicates()
print ("Shape of finalcombineddata after dropping duplicates", finalcombineddata.shape)
```

```python
duplicate_rows_finalcombineddata_id = finalcombineddata[finalcombineddata.duplicated(['id'])]
print (duplicate_rows_finalcombineddata_id.shape)
```

```python
#Still missing values are present as NA
print (finalcombineddata.isna().sum())
print (finalcombineddata.isna().sum().sum())
```

```python
#Mean of normalised-losses are taken
avgnormalised = finalcombineddata[["normalised-losses"]].mean()
avgnormalised
```

```python
#NA values are being replaced with mean value of 'normalised-losses'
finalcombineddata[["normalised-losses"]] = finalcombineddata[["normalised-losses"]].replace(np.nan,avgnormalised)
```

```python
#Now no null values in 'normalised-losses'
print (finalcombineddata.isna().sum())
print (finalcombineddata.isna().sum().sum())
```

```python
#Same as normalised-losses here for 'price' it is done
avgprice = finalcombineddata[["price"]].mean()
avgprice
```

```python
finalcombineddata[["price"]] = finalcombineddata[["price"]].replace(np.nan,avgprice)
```

```python
#Now no null values in price column
print (finalcombineddata.isna().sum())
print (finalcombineddata.isna().sum().sum())
```

```python
#Continued to check for unique values in each column to find anomalies and replaced with correct values
#Here mix of both number and variable
finalcombineddata.iloc[:, 6].unique()
```

```python
#Here 'six' and 'sixth'
finalcombineddata.iloc[:, 16].unique()
```

```python
#diesel is two and having extra spaces
finalcombineddata.iloc[:, 4].unique()
```

```python
#Both mpfi and Mpfi are same
finalcombineddata.iloc[:, 18].unique()
```

```python
#Replacing all the anomalies with correct values
to_replace = {'two': '2','four': '4','five': '5','six': '6','three': '3','eight': '8','twelve': '12','4L': '4','2L': '2','diesel  ': 'diesel','Mpfi': 'mpfi'}
finalcombineddata.replace(to_replace, inplace=True)
```

# Univariate Analysis

```python
#Same as before, using value.count the bar graph is plotted for each column
ax = finalcombineddata['engine-type'].value_counts(normalize=False).plot(figsize=(7,4),kind='bar',color='red')
ax.set_xlabel("engine-type", fontsize=15, color='black')
ax.set_ylabel("count", fontsize=15, color='black')
plt.show()
```

```python
ax = finalcombineddata['make'].value_counts(normalize=False).plot(figsize=(12,6),kind='bar',color='green')
ax.set_xlabel("make", fontsize=15, color='black')
ax.set_ylabel("count", fontsize=15, color='black')
plt.show()
```

```python
ax = finalcombineddata['body-style'].value_counts(normalize=False).plot(figsize=(7,4),kind='bar',color='brown')
ax.set_xlabel("body-style", fontsize=15, color='black')
ax.set_ylabel("count", fontsize=15, color='black')
plt.show()
```
