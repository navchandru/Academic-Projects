```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Setting up the display of row and column to show all of the them
pd.set_option('display.max_row', 1000)
pd.set_option('display.max_columns', 50)
#loading all three data to dataframes using Pandas
dataframe1 = pd.read_csv("data1.csv")
dataframe2 = pd.read_csv("data2.csv")
dataframe3 = pd.read_csv("data3.csv")
```

```python
#Merged dataframe1 and dataframe2 based on 'id' column
combineddata = pd.merge(dataframe1, dataframe2, on = 'id')
finalcombineddata = pd.concat([combineddata, dataframe3])
finalcombineddata.to_csv("finalcombineddata.csv", index = False)
```

```python
#Mean of normalised-losses are taken
avgnormalised = finalcombineddata[["normalised-losses"]].mean()
avgnormalised
```

```python
#NA values are being replaced with mean value of 'normalised-losses'
finalcombineddata[["normalised-losses"]] = finalcombineddata[["normalised-losses"]].replace(np.nan,avgnormalised)
```

```python
#Now no null values in 'normalised-losses'
print (finalcombineddata.isna().sum())
print (finalcombineddata.isna().sum().sum())
```

```python
#Same as normalised-losses here for 'price' it is done
avgprice = finalcombineddata[["price"]].mean()
avgprice
```

```python
finalcombineddata[["price"]] = finalcombineddata[["price"]].replace(np.nan,avgprice)
```

```python
#Now no null values in price column
print (finalcombineddata.isna().sum())
print (finalcombineddata.isna().sum().sum())
```

# Bivariate analysis

```python
#With groupby of make and price, two variable bar graph is plotted
#Jaguar being the costliest and 'bmw','porsche', 'mercedes-benz' are also costly.
groupmake_price = finalcombineddata[['make','price']]
make_avg = groupmake_price.groupby(['make'], as_index=False).mean().sort_values("price", ascending=True)
make_avg
plt.figure(figsize=(16, 8))
plt.xticks(rotation = 90)
plt.bar("make", "price", data = make_avg, color = "blue")
plt.xlabel('make', fontsize=15, color='black')
plt.ylabel('price', fontsize=15, color='black')
plt.show()
```

```python
groupdrive_price = finalcombineddata[['drive-wheels','price']]
drive_avg = groupdrive_price.groupby(['drive-wheels'], as_index=False).mean().sort_values("price", ascending=False)
drive_avg
plt.figure(figsize=(12, 6))
plt.xticks(rotation = 90)
plt.bar("drive-wheels", "price", data = drive_avg, color = "blue")
plt.xlabel('drive-wheels', fontsize=15, color='black')
plt.ylabel('price', fontsize=15, color='black')
plt.show()
```

```python
#Have compared the engine type and engine size - as engine-size affects the price(from correlation plot)
groupenginetype_size = finalcombineddata[['engine-type','engine-size']]
enginetype_avg = groupenginetype_size.groupby(['engine-type'], as_index=False).mean().sort_values("engine-size", ascending=False)
enginetype_avg
plt.figure(figsize=(12, 6))
plt.xticks(rotation = 90)
plt.bar("engine-type", "engine-size", data = enginetype_avg, color = "blue")
plt.xlabel('engine-type', fontsize=15, color='black')
plt.ylabel('engine-size', fontsize=15, color='black')
plt.show()
```

```python
#Scatter plot with regression of engine-size and price. It shows that they are positively correlated and as size increases price increases
plt.figure(figsize=(12, 8))
sns.regplot(x=finalcombineddata["engine-size"], y=finalcombineddata["price"], color="g")
plt.show()
```

```python
#Have compared the body style of the car and price - as body-style is related to curb-weight and curb-weight is related to price(from the correlation plot)
groupbody_price = finalcombineddata[['body-style','price']]
body_avg = groupbody_price.groupby(['body-style'], as_index=False).mean().sort_values("price", ascending=True)
body_avg
plt.figure(figsize=(12, 6))
plt.xticks(rotation = 90)
plt.bar("body-style", "price", data = body_avg, color = "green")
plt.xlabel('body-style', fontsize=15, color='black')
plt.ylabel('price', fontsize=15, color='black')
plt.show()
```

```python
#Have compared the body style and curb-weight as style increases or decreases weight
groupbody_curb = finalcombineddata[['body-style','curb-weight']]
bodyc_avg = groupbody_curb.groupby(['body-style'], as_index=False).mean().sort_values("curb-weight", ascending=False)
bodyc_avg
plt.figure(figsize=(12, 6))
plt.xticks(rotation = 90)
plt.bar("body-style", "curb-weight", data = bodyc_avg, color = "blue")
plt.xlabel('body-style', fontsize=15, color='black')
plt.ylabel('curb-weight', fontsize=15, color='black')
plt.show()
```

```python
#Curb-weight and price are positively correlated and as curb-weight increases the price also increases
plt.figure(figsize=(12, 8))
sns.regplot(x=finalcombineddata["curb-weight"], y=finalcombineddata["price"], color="g")
plt.show()
```

```python
#Have compared the engine-type and curb-weight
#dohcv and ohcv has more curb weight compared to other types
groupenginet_curb = finalcombineddata[['engine-type','curb-weight']]
enginet_avg = groupenginet_curb.groupby(['engine-type'], as_index=False).mean().sort_values("curb-weight", ascending=False)
enginet_avg
plt.figure(figsize=(12, 6))
plt.xticks(rotation = 90)
plt.bar("engine-type", "curb-weight", data = enginet_avg, color = "blue")
plt.xlabel('engine-type', fontsize=15, color='black')
plt.ylabel('curb-weight', fontsize=15, color='black')
plt.show()
```
